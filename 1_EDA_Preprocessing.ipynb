## PREPROCESSING STEPS

Because the reads are measured only as relative abundances, we must use the QPCR measurements to convert them to absolute measures (which are comparable across time). To do this, we simply calculate the proportion (relative / total) of each bug and then multiply by QPCR.

```python
def preprocess_mouse(mouse_reads_df, mouse_qpcr):
  mouse_reads = mouse_reads_df.values
  mouse_qpcr = mouse_qpcr.values
  
  mouse_read_totals = np.sum(mouse_reads, axis=0)
  mouse_read_percent = mouse_reads * 1/mouse_read_totals
  
  mouse_read_check = np.sum(mouse_read_percent, axis=0)
  assert np.sum(mouse_read_check) == mouse_reads.shape[1], 'mouse reads proportions didnt sum to one'
  
  mouse_read_absolute_mean = mouse_read_percent * mouse_qpcr[:,0]
  mouse_read_absolute_std = mouse_read_percent * mouse_qpcr[:,1]
  
  return mouse_read_absolute_mean, mouse_read_absolute_std
```

```python
fig, ax = plt.subplots(3,3,figsize=(20,15))
ax = ax.ravel()

for mouse in range(2,11):
  mouse_reads, mouse_qpcr, mouse_times = get_mouse_data(mouse, 'reads'), get_mouse_data(mouse, 'qpcr'), get_mouse_data(mouse, 'times')  
  mouse_reads_abs_mu, mouse_reads_abs_std = preprocess_mouse(mouse_reads, mouse_qpcr)
  ax[mouse-2].plot(mouse_times, np.sum(mouse_reads_abs_mu, axis=0), label='Total')
  for i in range(20):
    ax[mouse-2].plot(mouse_times, mouse_reads_abs_mu[i,:], label='OTU_{}'.format(i))
    ax[mouse-2].plot(mouse_times, mouse_reads_abs_mu[i,:], label='OTU_{}'.format(i))
    ax[mouse-2].plot(mouse_times, mouse_reads_abs_mu[i,:], label='OTU_{}'.format(i))
  ax[mouse-2].set_title('Mouse = {}'.format(mouse))
  
```
